```python
#Ejercicio 3
# Método	               Operador	   Qué hace
# union()	                  |	       Junta todo, eliminando duplicados.
# intersection()	          &	       Solo lo que está en ambos conjuntos.
# difference()	              -	       Lo que está en el A pero no en el B.
# symmetric_difference()	  ^	       Lo que está en uno o en otro, pero no en ambos.

set1 = {1,2,3,4,5}
set2 = {2,4,6,8,10}
print(f"Set 1 {set1}")
print(f"Set 2 {set2}")
print("#########################")
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
print(set1.symmetric_difference(set2))
print("#########################")
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)
print(set1 ^ set2)
```

```python
#Ejercicio 12
#Metodos de sets
# Método	¿Lanza error?    	¿Devuelve el elemento borrado?	Característica clave
# remove	Sí	                               No	            Estricto.
# discard	No	                               No	            Seguro/Silencioso.
# pop	    Sí (si está vacío)	               Sí	            Aleatorio.
# clear	    N/A	                               No	            Vacía todo el contenedor.

set1={1,2,3,4,5}
print(set1)
for i in range(0,len(set1)):
    set1.pop()

print(set1)

set2={1,2,3,4,5}
print(set2)

while set2:
    set2.pop()

print(set2)
```

```python
#Ejercicio 14
#Recordar que in funciona en varias estructuras de datos
# Estructura	Ejemplo	                   Comportamiento
# Listas	      11 in [10, 11, 12]	   True. Busca el valor exacto.
# Tuplas	      11 in (10, 11)	       True. Igual que las listas.
# Strings	     "py" in "python"	       True. Busca subcadenas.
# Diccionarios	 "nombre" in mi_dict	   True si la clave existe (no busca en los valores).
# Ranges	      5 in range(1, 10)	       True. Muy eficiente matemáticamente.

set1={1,2,3,4,5,11,12,13}

print(f"2 esta en el set {2 in set1}")

print(f"11 esta en el set {11 in set1}")
```
