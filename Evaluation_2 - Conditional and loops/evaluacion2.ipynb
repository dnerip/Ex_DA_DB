```python
## Programa 13

# Este programa aprovecha que la sentencia input se recibe 
# como cadena para posteriormente utilizar el string 
# como arreglo y sumar los digitos casteandolos a int.

num = input("Ingrese un numero: ")

i = 0
sum = 0

while i <= len(num)-1:
    sum = sum + int(num[i])  
    i += 1

print(sum)
```
