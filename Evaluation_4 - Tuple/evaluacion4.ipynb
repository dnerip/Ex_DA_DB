```python
#Ejercicio 15
# Existen varias funciones en python para cumplir con la tarea
# min() , max(), sum() son ejemplos de estas funcionan sobre iterables
# len() -> longitud de un iterable o cadena
# any() -> devuelve true si al menos uno de los elementos es verdadero
# all() -> devuelve true si todos sus elementos son verdaderos
# sorted() -> devuelve una ñista con los elementos ordenados
# enumerate() -> te da el inidice y el valor al mismo tiempo en un bucle
# Ej. lista = list(enumerate(tupla))

# zip() -> combina dos o mas iterables elemento a elemento

#recuerda que  para min(), max() y sorted() aceptan keys que define
#el criterio por el cual deseas comparar
#Ej. 
#    palabras = ["python", "ia", "programacion"]
#    mejor_palabra = max(palabras, key=len) # 
#    
#    Resultado: "programacion"

def minmaxsum(tupla):
    return min(tupla), max(tupla), sum(tupla)

tupla = tuple(range(1,101))

print(minmaxsum(tupla))
```

```python
#recuerda que  para min(), max() y sorted() aceptan keys que define
#el criterio por el cual deseas comparar
#Ej. 
#    palabras = ["python", "ia", "programacion"]
#    mejor_palabra = max(palabras, key=len) # 
#    
#    Resultado: "programacion"

#Escenario               |  Función Key Sugerida
#Cadenas de texto        | "len, str.lower, str.upper"
#Números absolutos,      | "abs (para que -10 sea ""mayor"" que 5)"
#Listas de diccionarios, | lambda x: x['campo'] o itemgetter('campo')
#Objetos propios         | lambda x: x.atributo o attrgetter('atributo')

###################################################################################
#Basado en la longitud
nombres = ["Ana", "Sebastian", "Zoe", "Max"]

# El nombre más largo
print(max(nombres, key=len))  # Resultado: "Sebastian"

# Ordenar por longitud de menor a mayor
print(sorted(nombres, key=len)) # Resultado: ['Ana', 'Zoe', 'Max', 'Sebastian']

###################################################################################
#Ignorar mayusculas y minusculas
letras = ["a", "B", "c", "A"]

# Sin key: las mayúsculas ganan (A=65, a=97)
print(min(letras)) # Resultado: "A"

# Con key: ignora si es mayúscula o minúscula
print(min(letras, key=str.lower)) # Resultado: "a" (o la primera "A" que encuentre)

###################################################################################
#Trabajando con Diccionarios o Tuplas

carrito = [
    {"producto": "Libro", "precio": 15},
    {"producto": "Café", "precio": 5},
    {"producto": "Monitor", "precio": 200}
]

# Buscamos el diccionario que tenga el "precio" más bajo
barato = min(carrito, key=lambda item: item["precio"])

print(barato) # Resultado: {'producto': 'Café', 'precio': 5}


###################################################################################
# Usando itemgetter (Más rápido que lambda)

from operator import itemgetter

estudiantes = [("Juan", 18), ("Maria", 22), ("Luis", 20)]

# Ordenar por la edad (índice 1 de la tupla)
jovenes_primero = sorted(estudiantes, key=itemgetter(1))

print(jovenes_primero) # Resultado: [('Juan', 18), ('Luis', 20), ('Maria', 22)]



```
