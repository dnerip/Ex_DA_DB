**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here

#En este ejemplo se utiliza la recursividad en funciones,
#Se llama a la misma funcion para ir obteniendo los factoriales desde la base

def factorial(num):
    print("Entro al fac")
    if num == 0:
        return 1
    else:
        print(num)
        return num * factorial(num-1)
    
num = int(input("Ingrese un numero para calcular el factorial: "))
print("El factorial del numero " + str(num) + " es " + str(factorial(num)))

# OTRA FORMA DE MOSTRAR EL RESULTADO CON LA SALIDA FORMATEADA

print(f"El factorial de {num} es {factorial(num)}")

    
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
# Your code here
num = int(input("Ingrese un numero para obtener la suma de los primeros numeros naturales: "))

#Solucion posible mediante un ciclo while

i=num
suma = 0
while i!=0:
    suma = suma + i
    i = i-1

print(suma)
    
#La siguiente solucion se le atribuye al matemático Carl Friedrich Gauss, 
#quien, según la leyenda, la descubrió cuando tenía apenas 10 años para 
#resolver un castigo de su profesor.

suma = (num * (num+1)) // 2

print(f"La suma de los numeros naturales es: {suma}")
```

**Question 17:** Write a Python program to check if a year is a leap year.

**Question 18:** Write a Python program to reverse a string.

```python
# Your code here

cad = input("Ingrese la cadena a invertir: ")
invertida = ""
for caracter in cad:
    #Vamos recorriendo los caracteres uno por uno
    #en la concatenacion esta la clave
    invertida = caracter + invertida

print(invertida)

#manera alternativa usando Slicing [::], la opción -1
#le indica a python el paso tomando desde el ultimo caracter
#y los dos primeros espacios vacios le dice que queremos toda la cadena
inv = cad[::-1]
print(inv)
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here

pal = input("Ingrese la pabra a verificar: ")
temp = pal[::-1]
if pal == temp:
    print(f"La palabra {temp} es un palindromo ")
else: 
    print(f"La palabra {temp} no es un palindromo ")
```

**Question 20:** Write a Python program to sort a list of numbers in ascending order.

```python
# Your code here

#Recordar que se pueden crear arreglos desde parametros, en este caso desde un input
#haciendo uso de .split para entender que los numeros estan separados por un espacio
#se realiza el ciclo for para recorrer dichos elementos del arreglo ingresado en el input
#y el formateo a la variable pivote donde se van almacenando los elementos para que 
#la instruccion ordene por el valor numerico y no el alfabetico ya que recordemos que
#input almacena en la variable que se le indique como una cadena

array = [int(num) for num in input("Ingrese los numeros separados por un espacio: ").split()]
array.sort()
print(array)
```
