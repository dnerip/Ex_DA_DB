```python
#Ejercicio 11

dict1 = {i: i**2 for i in range(1,11)}
dict2 = {}

for k,i in dict1.items():
    if k%2 == 0:
        #dict2.update({k:i})
        dict2 = dict2 | {k:i}

print(dict2)
```

```python
#Ejercicio 12

dict1 = {i: i**2 for i in range(1,6)}
dict2 = {}

for k,i in dict1.items():
    dict2 = dict2 | {i:k}

print(dict2)
```

```python
#Ejercicio 14

def count(cadena):
    new = {}
    for i in cadena:
        new[i] = new.get(i,0)+1

    return new

inp = input("Ingrese la cadena: ")

print(count(inp))
```
